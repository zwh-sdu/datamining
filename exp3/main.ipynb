```python
import math
import numpy as np
import functools

qrels_dict = {}#字典用来存储输入的文件qrels.txt
qrels_real = {}#计算NDCG中要计算IDCG的ideal_score的字典
q_num=0 #查询个数
qrels_file = 'qrels2014.txt'
result_file_lncltn = 'result_lncltn.txt'
result_file_nncntn = 'result_nncntn.txt'
result_file_ancatn = 'result_ancatn.txt'
```

```python
qrels_dict = {}
with open(qrels_file, 'r') as f:
    for line in f:
        t = line.strip().split(' ')
        if t[0] not in qrels_dict:
            qrels_dict[t[0]] = {}
        if int(t[3]) > 0:
            qrels_dict[t[0]][t[2]] = int(t[3])
```

```python
qrels_dict
```

```python
result_dict_lncltn = {}
result_dict_nncntn = {}
result_dict_ancatn = {}
with open(result_file_lncltn, 'r') as f:
    for line in f:
        t = line.strip().split(',')
        if t[0] not in result_dict_lncltn:
            result_dict_lncltn[t[0]] = []
        result_dict_lncltn[t[0]].append(t[1])
with open(result_file_nncntn, 'r') as f:
    for line in f:
        t = line.strip().split(',')
        if t[0] not in result_dict_nncntn:
            result_dict_nncntn[t[0]] = []
        result_dict_nncntn[t[0]].append(t[1])
with open(result_file_ancatn, 'r') as f:
    for line in f:
        t = line.strip().split(',')
        if t[0] not in result_dict_ancatn:
            result_dict_ancatn[t[0]] = []
        result_dict_ancatn[t[0]].append(t[1])
```

## MAP

```python
AP_lncltn = [] #存储每一个query的average precision
AP_nncntn = []
AP_ancatn = []
for q in result_dict_lncltn:
    result_q = result_dict_lncltn[q] #查询q得到的结果
    true_q = set(qrels_dict[q].keys()) #对于q真正相关的docid的集合
    precision = []
    i=0
    yes=0
    for docid in result_q:
        i+=1
        if(docid in true_q):
            yes+=1
            precision.append(yes/i)
    if yes:
        AP_lncltn.append(np.sum(precision)/yes)
MAP_lncltn = np.mean(AP_lncltn)
for q in result_dict_nncntn:
    result_q = result_dict_nncntn[q] #查询q得到的结果
    true_q = set(qrels_dict[q].keys()) #对于q真正相关的docid的集合
    precision = []
    i=0
    yes=0
    for docid in result_q:
        i+=1
        if(docid in true_q):
            yes+=1
            precision.append(yes/i)
    if yes:
        AP_nncntn.append(np.sum(precision)/yes)
MAP_nncntn = np.mean(AP_nncntn)
for q in result_dict_ancatn:
    result_q = result_dict_ancatn[q] #查询q得到的结果
    true_q = set(qrels_dict[q].keys()) #对于q真正相关的docid的集合
    precision = []
    i=0
    yes=0
    for docid in result_q:
        i+=1
        if(docid in true_q):
            yes+=1
            precision.append(yes/i)
    if yes:
        AP_ancatn.append(np.sum(precision)/yes)
MAP_ancatn = np.mean(AP_ancatn)
```

```python
print("lnc.ltn方法对两个query的查询的AP得分为：")
print(AP_lncltn)
print("nnc.ntn方法对两个query的查询的AP得分为：")
print(AP_nncntn)
print("anc.atn方法对两个query的查询的AP得分为：")
print(AP_ancatn)
print("lnc.ltn方法的MAP得分为：")
print(MAP_lncltn)
print("nnc.ntn方法的MAP得分为：")
print(MAP_nncntn)
print("anc.atn方法的MAP得分为：")
print(MAP_ancatn)
```

## MRR

```python
RR_lncltn = []
RR_nncntn = []
RR_ancatn = []
for q in result_dict_lncltn:
    result_q = result_dict_lncltn[q] #查询q得到的结果
    true_q = set(qrels_dict[q].keys()) #对于q真正相关的docid的集合
    i = 0
    for docid in result_q:
        i+=1
        if docid in true_q:
            RR_lncltn.append(1/i)
            break
MRR_lncltn = np.mean(RR_lncltn)
for q in result_dict_nncntn:
    result_q = result_dict_nncntn[q] #查询q得到的结果
    true_q = set(qrels_dict[q].keys()) #对于q真正相关的docid的集合
    i = 0
    for docid in result_q:
        i+=1
        if docid in true_q:
            RR_nncntn.append(1/i)
            break
MRR_nncntn = np.mean(RR_nncntn)
for q in result_dict_ancatn:
    result_q = result_dict_ancatn[q] #查询q得到的结果
    true_q = set(qrels_dict[q].keys()) #对于q真正相关的docid的集合
    i = 0
    for docid in result_q:
        i+=1
        if docid in true_q:
            RR_ancatn.append(1/i)
            break
MRR_ancatn = np.mean(RR_ancatn)
```

```python
print("lnc.ltn方法对两个query的查询的RR得分为：")
print(RR_lncltn)
print("nnc.ntn方法对两个query的查询的RR得分为：")
print(RR_nncntn)
print("anc.atn方法对两个query的查询的RR得分为：")
print(RR_ancatn)
print("lnc.ltn方法的MRR得分为：")
print(MRR_lncltn)
print("nnc.ntn方法的MRR得分为：")
print(MRR_nncntn)
print("anc.atn方法的MRR得分为：")
print(MRR_ancatn)
```

## NDCG

```python
NDCG_each_lncltn = []
NDCG_each_nncntn = []
NDCG_each_ancatn = []
for q in result_dict_lncltn:
    result_q = result_dict_lncltn[q]
    true_q = list(qrels_dict[q].values())
    idea_list = sorted(true_q, reverse=True) #理想情况下检索结果（按照relevance从大到小）
    i = 1
    DCG = qrels_dict[q].get(result_q[0], 0) #在每一个CG的结果上除以一个折损值
    IDCG = idea_list[0]
    length =len(idea_list)
    for docid in result_q[1: length]:
        i += 1
        rel = qrels_dict[q].get(docid, 0)  #.get后面两个参数，表示如果找不到docid，就返回0
        DCG += rel / math.log(i, 2)
        IDCG += idea_list[i-1] / math.log(i, 2)
    NDCG = DCG / IDCG #归一化处理
    NDCG_each_lncltn.append(NDCG)
NDCG_lncltn = np.mean(NDCG_each_lncltn)
for q in result_dict_nncntn:
    result_q = result_dict_nncntn[q]
    true_q = list(qrels_dict[q].values())
    idea_list = sorted(true_q, reverse=True)
    i = 1
    DCG = qrels_dict[q].get(result_q[0], 0) #在每一个CG的结果上除以一个折损值
    IDCG = idea_list[0]
    length =len(idea_list)
    for docid in result_q[1: length]:
        i += 1
        rel = qrels_dict[q].get(docid, 0)  #.get后面两个参数，表示如果找不到docid，就返回0
        DCG += rel / math.log(i, 2)
        IDCG += idea_list[i-1] / math.log(i, 2)
    NDCG = DCG / IDCG #归一化处理
    NDCG_each_nncntn.append(NDCG)
NDCG_nncntn = np.mean(NDCG_each_nncntn)
for q in result_dict_ancatn:
    result_q = result_dict_ancatn[q]
    true_q = list(qrels_dict[q].values())
    idea_list = sorted(true_q, reverse=True)
    i = 1
    DCG = qrels_dict[q].get(result_q[0], 0) #在每一个CG的结果上除以一个折损值
    IDCG = idea_list[0]
    length =len(idea_list)
    for docid in result_q[1: length]:
        i += 1
        rel = qrels_dict[q].get(docid, 0)  #.get后面两个参数，表示如果找不到docid，就返回0
        DCG += rel / math.log(i, 2)
        IDCG += idea_list[i-1] / math.log(i, 2)
    NDCG = DCG / IDCG #归一化处理
    NDCG_each_ancatn.append(NDCG)
NDCG_ancatn = np.mean(NDCG_each_ancatn)
```

```python
print("lnc.ltn方法对两个query的查询的NDCG得分为：")
print(NDCG_each_lncltn)
print("nnc.ntn方法对两个query的查询的NDCG得分为：")
print(NDCG_each_nncntn)
print("anc.atn方法对两个query的查询的NDCG得分为：")
print(NDCG_each_ancatn)
print("lnc.ltn方法的NDCG得分为：")
print(NDCG_lncltn)
print("nnc.ntn方法的NDCG得分为：")
print(NDCG_nncntn)
print("anc.atn方法的NDCG得分为：")
print(NDCG_ancatn)
```
